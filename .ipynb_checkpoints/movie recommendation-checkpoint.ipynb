```python
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
selected_columns = ['genres' , 'keywords','tagline','cast','director']
selected_columns
```

```python
for feature in selected_columns:
    df[feature] = df[feature].fillna('')
```

```python
combined_feature = ''
for feature in selected_columns:
    combined_feature = combined_feature + df[feature] +' '
combined_feature
```

```python
vectorizer = TfidfVectorizer()
```

```python
vect_feature = vectorizer.fit_transform(combined_feature)
```

```python
print(vect_feature)
```

```python
similarity = cosine_similarity(vect_feature)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
name  = input("Enter your Facourite Movie name:")
```

```python
all_movies = df['title'].tolist() 
```

```python
all_movies
```

```python
closest_match = difflib.get_close_matches(name , all_movies)
```

```python
print(closest_match)
```

```python
closest_match = closest_match[0]
closest_match
```

```python
index = df[df.title == closest_match]['index'].values[0]
index
```

```python
similarity_score = list(enumerate(similarity[index]))
```

```python
similarity_score
```

```python
len(similarity_score)
```

```python
sorted_similarity_score = sorted(similarity_score , key = lambda x:x[1] , reverse=True)
```

```python
print(sorted_similarity_score)
```

```python
print("Movies Suggested based on your Fav:")
i = 1
for movies in sorted_similarity_score:
    index = movies[0]
    title  = df[df.index == index]['title'].values[0]
    if(i<31):
        print(i, '.' , title)
        i+=1
```

```python
name  = input("Enter your Facourite Movie name:")

all_movies = df['title'].tolist() 

closest_match = difflib.get_close_matches(name , all_movies)
closest_match = closest_match[0]

index = df[df.title == closest_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index]))

sorted_similarity_score = sorted(similarity_score , key = lambda x:x[1] , reverse=True)

print("Movies Suggested based on your Fav:")
i = 1
for movies in sorted_similarity_score:
    index = movies[0]
    title  = df[df.index == index]['title'].values[0]
    if(i<31):
        print(i, '.' , title)
        i+=1
```
